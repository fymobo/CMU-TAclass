```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
plt.style.use('classic')
x = np.linspace(0,10,100)

fig = plt.figure(dpi=300)
plt.plot(x, np.sin(x),'p')  #
plt.plot(x,np.cos(x),'--')
plt.show()
```

```python
fig.savefig('my_figure.png')
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))
plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
```
